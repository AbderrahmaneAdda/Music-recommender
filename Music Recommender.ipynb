```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
```

```python
# load the dataset7
df = pd.read_csv('top10s.csv', encoding = "ISO-8859-1")
df = df.drop(['Unnamed: 0', '', ''], axis=1)
df.head()
```

```python
#On va enlever les colonnes que l'on va pas utiliser , on va garder :  artist, genre, year, bpm, dur, spc, pop
#Avant ça simplifions les genre
df.topgenre.unique()
```

```python
df1 = df.drop(['title', 'nrgy', 'dnce', 'dB','live','val', 'acous'], axis=1)
df1.head()
```

```python
#Corrigons tout cela : 

df1.loc[df['topgenre'] == 'pop', 'topgenre'] = 'pop'
df1.loc[df['topgenre'] == 'neo mellow', 'topgenre'] = 'pop'

df1.loc[df['topgenre'] == 'hip hop', 'topgenre'] = 'hip hop'
df1.loc[df['topgenre'] == 'rap', 'topgenre'] = 'hip hop'
df1.loc[df['topgenre'] == 'hollywood', 'topgenre'] = 'hip hop'

df1.loc[df['topgenre'] == 'electro', 'topgenre'] = 'electro'
df1.loc[df['topgenre'] == 'electronic', 'topgenre'] = 'electro'
df1.loc[df['topgenre'] == 'electropop', 'topgenre'] = 'electro'
df1.loc[df['topgenre'] == 'edm', 'topgenre'] = 'electro'
df1.loc[df['topgenre'] == 'downtempo', 'topgenre'] = 'electro'
df1.loc[df['topgenre'] == 'brostep', 'topgenre'] = 'electro'
df1.loc[df['topgenre'] == 'complextro', 'topgenre'] = 'electro'

df1.loc[df['topgenre'] == 'house', 'topgenre'] = 'house'
df1.loc[df['topgenre'] == 'big room', 'topgenre'] = 'house'

df1.loc[df['topgenre'] == 'rock', 'topgenre'] = 'rock'
df1.loc[df['topgenre'] == 'permanent wave', 'topgenre'] = 'rock'
```

```python
s=0
l=['pop','hip hop','electro','house', 'rock']
for i in df1['topgenre']:
    if i not in l:
        df1 = df1.drop([s])
    s=s+1
```

```python
df1
#On a simplifier les genres et supprimé les colonnes inutiles
#On doit maintenant transformer les noms d'artiste en variables numeriques
```

```python
df11 = pd.get_dummies(df1, columns=["artist"])
```

```python
for i in df11.index:
    a = input('Do you like : ' + df['title'][i] + ' de ' + df['artist'][i] )
    if a == 'oui':
        df11.loc[i, 'Like'] = 1
    else:
        df11.loc[i, 'Like'] = 0
    
```

```python
df11 = pd.get_dummies(df1, columns=['topgenre', 'artist' ])
```

```python
# split into input (X) and output (y) variables
X = df11.iloc[:,0:49].values
y = df11.iloc[:,49].values

# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=49, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
